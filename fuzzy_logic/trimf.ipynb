# Funciones de membresía 

## Función de membresía triangular

```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.arange(60, 120, 0.1)
```

```python
# Definición de las funciones de membresía triangular
slow = fuzz.trimf(x, [40, 60, 80])
medium = fuzz.trimf(x, [70, 90, 100])
medium_fast = fuzz.trimf(x, [90, 100, 110])
full_speed = fuzz.trimf(x, [100, 120, 120])
```

```python
plt.figure()
plt.plot(x, full_speed, 'b', linewidth=1.5, label='Full Speed') 
plt.plot(x, medium_fast, 'k', linewidth=1.5, label='Medium Fast') 
plt.plot(x, medium, 'm', linewidth=1.5, label='Medium Powered') 
plt.plot(x, slow, 'r', linewidth=1.5, label='Slow') 
plt.title('Penalty Kick Fuzzy')
plt.ylabel('Membership')
plt.xlabel("Speed (Miles Per Hour)")
plt.legend(loc='center right', bbox_to_anchor=(1.25, 0.5), ncol=1, fancybox=True, shadow=True)
```

## Función de membresía trapezoidal

```python
x = np.arange(30, 130, 0.1)

slow = fuzz.trapmf(x, [45, 55, 60, 70])
medium = fuzz.trapmf(x, [60, 70, 80, 90])
medium_fast = fuzz.trapmf(x, [80, 90, 100, 105])
full_speed = fuzz.trapmf(x, [100, 110, 120, 120])
```

```python
plt.figure()
plt.plot(x, full_speed, 'b', linewidth=1.5, label='Full Speed') 
plt.plot(x, medium_fast, 'k', linewidth=1.5, label='Medium Fast') 
plt.plot(x, medium, 'm', linewidth=1.5, label='Medium Powered') 
plt.plot(x, slow, 'r', linewidth=1.5, label='Slow') 
plt.title('Penalty Kick Fuzzy')
plt.ylabel('Membership')
plt.xlabel("Speed (Miles Per Hour)")
plt.legend(loc='center right', bbox_to_anchor=(1.25, 0.5), ncol=1, fancybox=True, shadow=True)
```
